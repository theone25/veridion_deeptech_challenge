```python
# import module
import requests
import pandas as pd
```

```python
links = pd.read_csv("../Data/furniture stores pages.csv",header=0)
```

there seems to be an issue with how requests handles cloudflare approach to data scraping, an alternative would be to use urllib and some custom headers

```python
def url_ok(url):
    try:
        response = requests.get(url)
        print(response.status_code)
        if response.status_code == 200:
            return True
        else:
            return False
    except requests.ConnectionError as e:
        return e
```

```python
for link in links["max(page)"]:
    if url_ok(link) == False:
        print(link)
```
